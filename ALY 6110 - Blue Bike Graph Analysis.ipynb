```
# Import Needed Packages
import pandas as pd # pandas is used for data manipulation
from zipfile import ZipFile # Zips the file from S3
import io # io is used for 
from urllib.request import urlopen # open csv files from url
import urllib.request;  # used to make the request
import numpy as np
from datetime import date
```

```
def PullData(begin_date, end_date):
    '''Function to pull data from S3 with a specified date range'''
    master_csv = pd.DataFrame()
    for i in range(begin_date,end_date):
        try:
            data_url = "https://s3.amazonaws.com/hubway-data/" + str(i) + "-bluebikes-tripdata.zip"
            r = urlopen(data_url).read()
            file = ZipFile(io.BytesIO(r))
            file.extractall
            url = str(i) + "-bluebikes-tripdata.csv"
            csv = file.open(url)
            csv = pd.read_csv(csv,sep = ',') # read_csv returns the dataframe type
            master_csv = pd.concat([master_csv, csv])
        except:
            " "
    return master_csv
bb_df = PullData(201901,202209)
bb_df
```
